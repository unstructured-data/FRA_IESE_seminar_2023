```python
# import libraries
from pypdf import PdfReader
```

```python
# load PDF
reader = PdfReader("test.pdf")
reader
```

```python
# get the number of pages
number_of_pages = len(reader.pages)
```

```python
# extract the text from one particular page (first page)
page = reader.pages[0]
text = page.extract_text()
print(text)
```

```python
# extract all the text
all_text = []
for page in reader.pages:
    all_text.append(page.extract_text())
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
# get the raw HTML of the whole website
my_url = "https://www.bankofengland.co.uk/monetary-policy-summary-and-minutes/2023/may-2023"
r = requests.get(my_url)
soup = BeautifulSoup(r.content,"html.parser")
#soup
```

```python
# select the <div> of our interest using its id
html_focus = soup.find("div", {"id": "output"})
type(html_focus)
```

```python
# extract all the text from the HTML
all_text = html_focus.get_text()
all_text
```
